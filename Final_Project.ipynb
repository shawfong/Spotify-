```python
import pandas as pd
import numpy as np

#data = np.loadtxt("spotify_dataset_V3.csv", delimiter = ";", skiprows = 1)
df = pd.read_csv("Halved_Spotify_Dataset_V3.csv", sep = ",", skiprows=1)

df_clean = df.drop("id", axis = "columns")
df_clean = df_clean.drop("Song URL", axis = "columns")
df_clean = df_clean.drop("Points (Total)", axis = "columns")
df_clean = df_clean.drop("Points (Ind for each Artist/Nat)", axis = "columns")
df_clean = df_clean.drop("# of Nationality", axis = "columns")
df_clean = df_clean.drop("# of Artist", axis = "columns")
df_clean = df_clean.drop("Date", axis = "columns")
df_clean = df_clean.drop("Nationality", axis = "columns")
df_clean = df_clean.drop("Continent", axis = "columns")

#get rid of the comma in Tyler's name because it messes up our .split()
#df_clean.replace("Tyler, The Creator", "Tyler The Creator", inplace = True)

# calculate popularity (# of songs on list / avg ranking of songs )

# count how many times artist shows up in values of Artist and divide by total number of songs 

# get list of artist names 
artist_names = df_clean['Artist (Ind.)'].unique()
#print("artist names", artist_names)

# count how many times artist shows up by creating a dictionary 
counts = df_clean['Artist (Ind.)'].value_counts()

# find total number of songs 
total_songs = len(df_clean)
print("total songs", total_songs)

popularity = {}

for artist in artist_names:
    count_artist = counts[artist]
    popularity[artist] = count_artist / total_songs

#print("artist dictionary", popularity)

# drop artist independent 
df_clean = df_clean.drop("Artist (Ind.)", axis = "columns")

# drop duplicates 
df_clean.drop_duplicates(inplace=True)


# replace artist with popularity score
for artist in df_clean['Artists']:
    split_artist = [a.strip() for a in artist.split(",")]
    print("split", split_artist)
    #scores = [popularity.get(key) for key in split_artist] 
    #scores = [popularity.get(key, 0.0) for key in split_artist]

    scores = [] 
    for key in split_artist:
        if (key in popularity): scores.append(popularity.get(key))
        else: 
            scores.append(0.0)  # For now, just use 0.0 for these cases

    print("scores", scores)
    avg_scores = sum(scores) / len(scores)
    print("avg", avg_scores)

    #add new dictionary entries for multiple artists
    if len(scores) > 1:
        popularity[artist] = avg_scores
    else: popularity[artist]= 0.0

#replace artist names with popularity scores
df_clean["Artists"] = df_clean["Artists"].replace(popularity)

#shuffle data
shuffled_df_clean = df_clean.sample(frac = 1)
copy_shuffled_df_clean = shuffled_df_clean.copy()

#drop song titles from the copy
copy_shuffled_df_clean = copy_shuffled_df_clean.drop("Title", axis = "columns")

#display(shuffled_df_clean[:10])

y = copy_shuffled_df_clean.pop("Rank").to_numpy()
X = copy_shuffled_df_clean.to_numpy()
#display(X[:10])
#display(y[:10])
```

```python
#Split data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, shuffle = False)
```

```python
#Feature Scaling

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#Multivariate Linear Regression

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)
LinRegModel = LinearRegression().fit(X_train_scaled, y_train)
prediction = LinRegModel.predict(X_train_scaled)
#prediction2 = LinRegModel.predict(y_test)
print('r2 score', r2_score(y_train, prediction))
print('mean squared', mean_squared_error(y_train,prediction))
print(y_train,y_test)
```
